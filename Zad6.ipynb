```
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn import metrics
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

raw_data = load_iris()
all_data=raw_data.copy()

data=all_data['data']
#labels
feature_names=all_data['feature_names']
#features
target=all_data['target']
target_names=all_data['target_names']

df = pd.DataFrame(data=np.c_[data, target], columns=feature_names + ['target'])
X=data[:, :2]
y=pd.to_numeric(df['target'], downcast='signed')
#split data to test and train sets
print(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4) 
#C - default 1.0 (float)
clf_linear = SVC(C=1.5, kernel='linear')
clf_linear.fit(X_train, y_train)
y_pred = clf_linear.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
# Plotting decision regions
plot_decision_regions(X_train, y_train.to_numpy(), clf=clf_linear, legend=2)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
#validation x and validation y
plt.title(f'{clf_linear.score(X_test, y_test)}')
plt.show()

clf_poly = SVC(C=3, kernel='poly')
clf_poly.fit(X_train, y_train)
y_pred = clf_poly.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
plot_decision_regions(X_train, y_train.to_numpy(), clf=clf_poly, legend=2)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title(f'{clf_poly.score(X_test, y_test)}')
plt.show()

clf_sigmoid = SVC(C=4, kernel='sigmoid')
clf_sigmoid.fit(X_train, y_train)
y_pred = clf_sigmoid.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
plot_decision_regions(X_train, y_train.to_numpy(), clf=clf_sigmoid, legend=2)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title(f'{clf_sigmoid.score(X_test, y_test)}')
plt.show()
```
